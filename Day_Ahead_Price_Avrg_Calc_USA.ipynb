```python
from google.colab import files
uploaded = files.upload()  # (multi-select)
```

```python
import pandas as pd
import numpy as np
import io

def calculate_average_multiple_files(uploaded_files):
    """
    Reads multiple CSV files, calculates the average of numeric values
    in columns B through U (20 columns), and returns the final average.
    """
    # Column indices for B to U (0-indexed: B=1, C=2, ... U=20)
    columns_to_average = list(range(1, 21))  # [1, 2, 3, ..., 20]
    column_names = list("BCDEFGHIJKLMNOPQRSTU")  # For display

    all_numeric_values = []
    file_summaries = []

    for filename, file_content in uploaded_files.items():
        print(f"\n Processing: {filename}")

        # Read the CSV file
        df = pd.read_csv(io.BytesIO(file_content), header=None)
        print(f"   Rows: {len(df)} | Columns found: {len(df.columns)}")

        file_values = []

        for col_idx, col_name in zip(columns_to_average, column_names):
            if col_idx < len(df.columns):
                col_data = df.iloc[:, col_idx]
                numeric_col = pd.to_numeric(col_data, errors='coerce')
                valid_values = numeric_col.dropna().tolist()
                file_values.extend(valid_values)
            else:
                print(f"   Column {col_name} not found in {filename}, skipping.")

        file_avg = np.mean(file_values) if file_values else 0
        print(f"   Numeric values found: {len(file_values)} | File Average: {file_avg:.4f}")

        all_numeric_values.extend(file_values)
        file_summaries.append({"File": filename, "Values Count": len(file_values), "File Average": round(file_avg, 4)})

    # Final overall average across all files
    overall_average = np.mean(all_numeric_values) if all_numeric_values else 0

    # Print summary table
    print("\n" + "="*60)
    print("SUMMARY")
    print("="*60)
    summary_df = pd.DataFrame(file_summaries)
    print(summary_df.to_string(index=False))
    print("="*60)
    print(f"Total files processed : {len(uploaded_files)}")
    print(f"Total numeric values  : {len(all_numeric_values)}")
    print(f"FINAL OVERALL AVERAGE : {overall_average:.4f}")
    print("="*60)

    return overall_average

# Run the function
final_average = calculate_average_multiple_files(uploaded)
```

```python
Currency_convert_2023= 1.18
Mwh_to_kwh=1000
Final_average_price= (final_average*Currency_convert_2023)/Mwh_to_kwh
print(Final_average_price)
```
